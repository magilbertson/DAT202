### We're going to start with a few quick examples using NumPy and then we are going to dive into Pandas to analyze a really cool dataset

### First, I'm going to create a simple NumPy array using random numbers in a bunch of lists.

```python
import random as r
r.seed(a=100) #seed just ensures the same random numbers each time
list_1 = [r.random(),r.random()] #Creates list of 2 random numbers between 0-1
list_2 = [r.random(),r.random()]
list_3 = [r.random(),r.random()]
list_4 = [r.random(),r.random()]
print(list_1)
print(list_2)
print(list_3)
print(list_4)
```

### Now I'm going to convert my lists to a NumPy array

```python
import numpy as np
my_numpy = np.array([list_1,list_2,list_3,list_4])
print(my_numpy) #appears to be truncating but isn't
print(my_numpy[0,0]) #printing a specific data point shows the whole number
print(type(my_numpy))
```

### Now let's do it again starting with a 1D list

```python
list_all = []
list_all.extend(list_1+list_2+list_3+list_4)
print(list_all)
my_2nd_numpy = np.array(list_all) #numpy array of shape 1,8
my_2nd_numpy.shape = (4,2) #numpy array of shape 4,2
print(my_2nd_numpy)
my_2nd_numpy.shape
```

### Let's pull out some simple stat data from some columns

```python
print(interest_data['Music'].sum())
print(interest_data['Music'].mean())
print(interest_data['Music'].median())
print(interest_data['Music'].max())
print(interest_data['Music'].min())
print(interest_data['Music'].std()) #standard deviation
print(interest_data['Internet usage'].nunique()) #Number of unique values for this column
print(interest_data.shape)
```

### Now I want to figure out which column has the maximum mean value (meaning the most average liked music type)

```python
# Here I'm going to determine which of the music columns has the maximum mean
print(interest_data.iloc[:,1:18].mean().max()) #iloc allows me to grab columns/rows based on index positions
# I can also grab the column name associated with that maximum mean column
print(interest_data.iloc[:,1:18].mean().idxmax())
```

## Topic 3: Continuing with Pandas

### Now we want to start pulling more specific pieces of data from our larger dataframe to look at them closer

### I want to create a separate dataframe for the columns that are movie category interests so I can dig in a bit

```python
movie_interest_data = interest_data.iloc[:,19:31] #all rows, columns index 19-31
print(movie_interest_data.shape)
movie_interest_data.head()
```

### Now I only want to see the movie interests of people who are overall interested in movies.  I'm going to create another dataframe with only rows where the person said they had an interest of 5 in movies

```python
movie_manic_interest_data = movie_interest_data[movie_interest_data.Movies >=5]
movie_manic_interest_data.shape
```

### Let's pull our same trick from earlier and have a look at the averages for each of the columns

```python
movie_manic_interest_data.mean()
```

### Now sort it decending so I can see which movie types are most liked by moviephiles


```python
movie_manic_interest_data.mean().sort_values(ascending = False)
```

### Now I want to see which kinds of movies are the most popular with people that don't like comedy

```python
movie_manic_interest_data_no_comedy = movie_manic_interest_data[movie_manic_interest_data.Comedy <=3]
movie_manic_interest_data_no_comedy.shape
```

```python
movie_manic_interest_data_no_comedy.mean()
```

### Now to see how much difference between comedy haters and the overall group

```python
movie_manic_interest_data_no_comedy.mean().subtract(movie_manic_interest_data.mean())
```

### Now I want to see it descending again


```python
movie_manic_interest_data_no_comedy.mean().subtract(movie_manic_interest_data.mean()).sort_values(ascending=False)
```

### As a proud lefty, I'd like to see if this data shows any interesting disparities between left/right handed people.

```python
lefty_data = interest_data.groupby('Left - right handed')
```

```python
lefty_data.mean()
```

```python
lefty_mean_frame = lefty_data.mean().reset_index()
```

```python
lefty_mean_frame
```

```python
lefty_mean_frame.set_index('Left - right handed',inplace=True)
```

```python
lefty_mean_frame.diff(-1).sort_values(by = 'left handed',axis=1)
```

### Finally, I'd like to create some simple visuals that allow me to understand my data better

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12,18)
ax = interest_data.iloc[:,1:20].plot.box(color = {'boxes':'DarkGreen','whiskers':'DarkOrange','medians':'Blue'},vert=False)
print('') #to stop annoying output of xticklabels data
```

```python
plt.rcParams['figure.figsize']=(18,4)
ax = interest_data.iloc[:,1:19].mean().plot.bar()
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45, ha = 'right')
print("")
```
